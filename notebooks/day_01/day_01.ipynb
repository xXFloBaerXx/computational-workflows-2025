```python
# now set -resume to the command. What did change? 

!nextflow run nf-core/differentialabundance -r 1.5.0 -profile test,docker --outdir results -resume
```

Check out the current directory. Next to the outdir you specified, what else has changed?

When the pipeline is runned with -resume, Nextflow skips tasks that were already completed successfully and reuses their outputs from the previous run. SO bettween the first and the second one each computation has to be done again, but not between the second and the third.

```python
# delete the work directory and run the pipeline again using -resume. What did change?
!rm -rf work
!nextflow run nf-core/differentialabundance -r 1.5.0 -profile test,docker --outdir results -resume
```

What changed? Deleting the work directory removes all temporary computation results, so the pipeline needs to execute each step again. With the resume option it can cache all intermediate files and therefore can skip a few tasks. If the work directory is deleted the cache is deleted, so the resume argument has no impact on the performance anymore.
